```python
import pandas as pd
import re
```

```python
pdf_df = pd.read_excel('excelfiles/testing_input.xlsx', sheet_name='Sheet1')
test_summary_df = pd.read_excel('excelfiles/test.xlsx', sheet_name='Summary_test')
```

```python
keyword_finance = ['inward remittance','network international','sdm deposit']
```

```python
debit_ir = []
credit_ir = []

debit_cd = []
credit_cd = []
for i in range(len(pdf_df)):
        for keyword in keyword_finance:
                if keyword == 'inward remittance' or keyword == 'network international':
                        if keyword.lower() in pdf_df['Narration'][i].lower():
                                debit_ir.append(pdf_df['Debit'][i])
                                credit_ir.append(pdf_df['Credit'][i])
                elif keyword == 'sdm deposit':
                        if keyword.lower() in pdf_df['Narration'][i].lower():
                                debit_cd.append(pdf_df['Debit'][i])
                                credit_cd.append(pdf_df['Credit'][i])
                        
        
```

```python
len(credit_cd)
```

```python
total_ir = sum(credit_ir) - sum(debit_ir)
```

```python
total_ir
```

```python
total_cd = sum(credit_cd) - sum(debit_cd)
```

```python
total_cd
```

```python
for i in range(len(test_summary_df)):
    if test_summary_df['Description'][i].lower() == 'inward remittance':
        test_summary_df['Emirates NBD-Classic Luxury-Main'][i] = total_ir
    elif test_summary_df['Description'][i].lower() == 'cash deposited':
        test_summary_df['Emirates NBD-Classic Luxury-Main'][i]= total_cd
        
```
