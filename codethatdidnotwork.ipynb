```python
def extract_pdf_data_as_dataframe(file_name): 
    
    extracted_df_dict = {}
    extracted_raw_data_from_pdf = camelot.read_pdf(file_name, flavor='stream', pages='all')
    for i in range(extracted_raw_data_from_pdf.n):
        key_name = str(i)+'_dataframe_pdf_extracted'
        extracted_df_dict[key_name] = extracted_raw_data_from_pdf[i].df
    
    return extracted_df_dict        
    
```
